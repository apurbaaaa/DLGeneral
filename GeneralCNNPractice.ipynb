```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras import layers, models, datasets
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
print(x_train.shape)
```

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3), input_shape = (32, 32, 3) , activation='relu', padding = 'same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu', padding = 'same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu', padding = 'same'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),

    layers.Dense(32, activation = 'relu'),
    layers.Dropout(0.35),
    layers.Dense(64, activation = 'relu'),
    layers.Dense(10, activation = 'softmax')
])
    
    
```

```python
train_data_gen = ImageDataGenerator(
    rescale= 1./255,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip= True
)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y_train = LabelEncoder().fit_transform(y_train)
y_test = LabelEncoder().fit_transform(y_test)
```

```python
train_data = train_data_gen.flow(x_train, y_train)
```

```python
train_data
```

```python
test_data_gen = ImageDataGenerator(
    rescale= 1./255
)
```

```python
test_data = train_data_gen.flow(x_test, y_test)
```

```python
train_data.dtype
```

```python
model.compile(
    optimizer= 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
model.fit(train_data, validation_data=test_data, epochs=5, verbose=True)
```
